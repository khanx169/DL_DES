```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sns.set_style("darkgrid")
sns.set_palette("pastel")
```

```python
high_certainty_path = 'data/sdss-galaxyzoo/high_certainty/merged_dataset.csv'
full_over_lap_path = 'data/sdss-galaxyzoo/des_overlap/merged_dataset.csv'
```

```python
high_certainty_df = pd.read_csv( high_certainty_path )
full_overlap_df = pd.read_csv( full_over_lap_path )
```

```python
high_certainty_df.shape, full_overlap_df.shape
```

```python
high_certainty_df.head()
```

```python
full_overlap_df.head()
```

# Visualize ra and dec range 

### 1. High certainty df 

```python
high_certainty_df.ra.min(), high_certainty_df.ra.max()
```

```python
high_certainty_df.dec.min(), high_certainty_df.dec.max()
```

```python
high_certainty_df.ra.hist(bins=100)
plt.xlabel('ra in degrees')
plt.ylabel('# of galaxies')
plt.title('RA distribution for "high certanty" dataset')
```

```python
high_certainty_df.dec.hist(bins=100)
plt.xlabel('dec in degrees')
plt.ylabel('# of galaxies')
plt.title('DEC distribution for "high certanty" dataset')
```

```python
plt.hist2d(high_certainty_df['ra'], high_certainty_df['dec'], bins=[200, 100], range=[[0, 360], [-90, 90]], cmap='hot_r')
plt.colorbar(label='Count')
plt.xlabel('ra')
plt.ylabel('dec')
```

### 2. Full overlap df 

```python
full_overlap_df.ra.min(), full_overlap_df.ra.max()
```

```python
full_overlap_df.dec.min(), full_overlap_df.dec.max()
```

```python
full_overlap_df.ra.hist(bins=100, color='k')
plt.xlabel('ra in degrees')
plt.ylabel('# of galaxies')
plt.title('RA distribution for "full overlap" dataset')
```

```python
full_overlap_df.dec.hist(bins=100, color='k')
plt.xlabel('dec in degrees')
plt.ylabel('# of galaxies')
plt.title('DEC distribution for "full overlap" dataset')
```

```python
plt.hist2d(full_overlap_df['ra'], full_overlap_df['dec'], bins=[200, 100], range=[[0, 360], [-90, 90]], cmap='hot_r')
plt.colorbar(label='Count')
plt.xlabel('ra')
plt.ylabel('dec')
```

### Compare

```python
sns.kdeplot(high_certainty_df.ra, shade=True, label='high certainty')
sns.kdeplot(full_overlap_df.ra, shade=True, label='full overlap')
plt.xlabel('ra in degrees')
plt.title('KDE plot comparing ra coverage')
```

```python
sns.kdeplot(high_certainty_df.dec, shade=True, label='high certainty')
sns.kdeplot(full_overlap_df.dec, shade=True, label='full overlap')
plt.xlabel('dec in degrees')
plt.title('KDE plot comparing dec coverage')
```

# Visualize Probablility ranges 

### 1. High certainty df

```python
spiral_probs = high_certainty_df[ high_certainty_df.P_CS_DEBIASED > high_certainty_df.P_EL_DEBIASED ].P_CS_DEBIASED
ellip_probs = high_certainty_df[ high_certainty_df.P_CS_DEBIASED < high_certainty_df.P_EL_DEBIASED ].P_EL_DEBIASED

spiral_probs = np.array( spiral_probs )
ellip_probs = np.array( ellip_probs )
```

```python
len(spiral_probs)
```

```python
len(ellip_probs)
```

```python
len(high_certainty_df)
```

```python
plt.hist([ellip_probs, spiral_probs], bins=30, range=(0, 1), label=['Elliptical', 'Spiral'])
plt.legend()
plt.xlabel('Probability')
plt.ylabel('Galaxy count')
plt.title('Prob dist for each type in high certainty dataset')
```

```python
sns.kdeplot(spiral_probs, shade=True, label='spiral')
sns.kdeplot(ellip_probs, shade=True, label='elliptical')
plt.xlabel('Probability')
plt.title('KDE plot for prob. dist. of each type in high certainty dataset')
```

```python
sns.violinplot(spiral_probs)
plt.xlabel('Probability')
plt.title('Violin plot for spirals in "high certainty" dataset')
```

```python
sns.violinplot(ellip_probs, color='violet')
plt.xlabel('Probability')
plt.title('Violin plot for elliptical in "high certainty" dataset')
```

```python
sns.kdeplot(high_certainty_df.P_CS_DEBIASED, shade=True, color='g')
plt.title('Full P_CS_DEBIASED distribution for "high certainty" dataset')
```

```python
sns.kdeplot(high_certainty_df.P_EL_DEBIASED, shade=True, color='r')
plt.title('Full P_EL_DEBIASED distribution for "high certainty" dataset')
```
