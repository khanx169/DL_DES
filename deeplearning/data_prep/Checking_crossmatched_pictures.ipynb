```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sns.set_style("darkgrid")
sns.set_palette("pastel")
```

```python
PATH = "/home/khan74/scratch/new_DL_DES/"
```

```python
HP_crossmatch_df = pd.read_csv(PATH + 'deeplearning/data/high_prob_crossmatch_test_set.csv')
FO_crossmatch_df = pd.read_csv(PATH + 'deeplearning/data/full_overlap_crossmatch_test_set.csv')
```

```python
HP_SDSS_test_data_dir = f'{PATH}deeplearning/data/HP_crossmatch_test/sdss/'
HP_DES_test_data_dir = f'{PATH}deeplearning/data/HP_crossmatch_test/des/'

FO_SDSS_test_data_dir = f'{PATH}deeplearning/data/FO_crossmatch_test/sdss/'
FO_DES_test_data_dir = f'{PATH}deeplearning/data/FO_crossmatch_test/des/'
```

# HP Pictures 

```python
HP_crossmatch_df.shape
```

```python
def plot_HP(start_index = 0):
    
    for i, index in enumerate( HP_crossmatch_df.index[start_index:start_index+5] ):
    
        row = HP_crossmatch_df.loc[index]

        SDSS_OBJID = row.SDSS_OBJID
        DES_OBJID = row.DES_COADD_OBJECT_ID

        SDSS_img_path = HP_SDSS_test_data_dir + 'test/' + str(SDSS_OBJID) + '.png'
        DES_img_path = HP_DES_test_data_dir + 'test/' + str(DES_OBJID) + '.png'

        SDSS_img = plt.imread( SDSS_img_path )
        DES_img = plt.imread( DES_img_path )

        plt.subplot(2,5,i+1)
        plt.axis('Off')
        plt.subplots_adjust(wspace=0.1,hspace=0.05)
        plt.imshow(SDSS_img)
        plt.title('SDSS')

        plt.subplot(2,5,i+6)
        plt.axis('Off')
        plt.subplots_adjust(wspace=0.1,hspace=0.05)
        plt.imshow(DES_img)
        plt.title('DES')
        
    plt.show()
```

```python
for i in range(1000,1100,5):
    plot_HP(i)
    print('--------------------')
```

```python
def plot_FO(start_index = 0):
    
    for i, index in enumerate( FO_crossmatch_df.index[start_index:start_index+5] ):
    
        row = FO_crossmatch_df.loc[index]

        SDSS_OBJID = row.SDSS_OBJID
        DES_OBJID = row.DES_COADD_OBJECT_ID

        SDSS_img_path = FO_SDSS_test_data_dir + 'test/' + str(SDSS_OBJID) + '.png'
        DES_img_path  = FO_DES_test_data_dir + 'test/' + str(DES_OBJID) + '.png'

        SDSS_img = plt.imread( SDSS_img_path )
        DES_img = plt.imread( DES_img_path )

        plt.subplot(2,5,i+1)
        plt.axis('Off')
        plt.subplots_adjust(wspace=0.1,hspace=0.05)
        plt.imshow(SDSS_img)
        plt.title('SDSS')

        plt.subplot(2,5,i+6)
        plt.axis('Off')
        plt.subplots_adjust(wspace=0.1,hspace=0.05)
        plt.imshow(DES_img)
        plt.title('DES')
        
    plt.show()
```
